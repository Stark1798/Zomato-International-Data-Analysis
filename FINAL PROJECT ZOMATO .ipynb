```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
z1=pd.read_csv('zomato.csv',encoding='latin1')
```

```python
z1
z1.head()
```

```python
z2=pd.read_csv('Country-Code.csv')
```

```python
z2
```

```python
z3=pd.merge(z1,z2)
```

```python
z3
z3.head()
```

```python
z3.Country.value_counts()
```

## Show the Percentage of All countries Restaurant having Online Delivery ?

```python
z3['Has Online delivery'].value_counts().plot(kind='pie',autopct='%i%%')
plt.show()
```

##  Which Top 5 Countries are the Highest Buisness makers for  Zomato ?

```python
z4=z3.Country.value_counts().index
```

```python
z5=z3.Country.value_counts().values
```

```python
plt.pie(z5[:5],labels=z4[:5],autopct='%i%%',explode=[0,1,1,1,1])
plt.show()
```

##  Which Country Restaurants Has the Highest Aggregate Rating ?

```python
z6=z3.sort_values(
...     by="Aggregate rating",
...     ascending=False
... )
```

```python
z7=z6.head(10)
```

```python
sns.barplot(data=z7,x='Aggregate rating',y='Country')
plt.show()
```

##  Plot the graph for the Top 15 Cusines having most no of table booking ?

```python
z21=pd.crosstab(z3['Cuisines'],z3['Has Table booking'])
```

```python
z22=z21.head(15)
```

```python
z22.plot(kind='bar')
plt.grid()
plt.show()
```

##  Plot the graph showing top 10 Cusines having Highest price range ?

```python
q5=z3.sort_values(
...     by="Price range",
...     ascending=False
... )
```

```python
q51=q5.head(10)
```

```python
sns.barplot(data=q51,x='Price range',y='Cuisines')
plt.show()
```

##  Show the relation between Top Cities and their Restaurants Rating Text According to Votes ?


```python
v1=z3.sort_values(
...     by="Votes",
...     ascending=False
... )
```

```python
v2=v1.head(20)
```

```python
pd.crosstab(v2['City'],v2['Rating text']).plot(kind='bar')
plt.ylabel('Frequency')
plt.grid()
plt.show()
```

##   Show the currency used in top 10 rated restaurents As per Ratings ?

```python
z81=z3.sort_values(
...     by="Aggregate rating",
...     ascending=False
... )
```

```python
z82=z81.head(10)
```

```python
pd.crosstab(z82['Country'],z82['Currency']).plot(kind='bar')
plt.grid()
plt.show()
```
